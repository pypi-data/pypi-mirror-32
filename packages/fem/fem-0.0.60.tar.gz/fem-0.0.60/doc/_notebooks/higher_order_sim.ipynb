```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import fem
```

```python
n, m, degs = 20, 3, [4]
par = fem.simulate.model_parameters(n, m, degs)
```

```python
x = fem.simulate.time_series(par, n, m, l=100)
```

```python
fig, ax = plt.subplots(min(10,n), 1, figsize=(16,4))
for i in range(min(10,n)):
    ax[i].plot(x[i, :100], 'k-')
    ax[i].axis('off')
```
