```python
import seaborn as sns
sns.set_style("white")

%pylab inline
import numpy as np
import pylab
from mpl_toolkits.mplot3d import Axes3D

from nengolib.stats import sphere

n_samples = 100000
p = sphere.sample(n_samples, 3, rng=np.random.RandomState(1))

pal = np.asarray(sns.cubehelix_palette(3))
colors = abs(np.dot(p, [1, 0, 0]))*len(pal)

fig = pylab.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.set_axis_off() 

ax.scatter(*p.T, alpha=0.6, s=3, lw=0, c=pal[np.asarray(colors, dtype=int)])
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)
ax.view_init(elev=60, azim=48)

fig.tight_layout()
pylab.show()

fig.savefig('_sphere.png', bbox_inches='tight', pad_inches=-1.1, dpi=300)
```
