```python
%matplotlib inline
import matplotlib.pyplot as plt
from __future__ import division
import numpy as np
from scipy import stats, integrate, constants, optimize, interpolate
import _resolution_kernel as rk
from refnx.util import general
```

```python
pdf = rk.P_Wavelength(358, 7500, 24, 4, da=0.005)
pdfa = rk.P_Theta(1.611, 0.167)

wave = 13

mean_q = general.q(0.65, wave)
max_q = general.q(0.65 + 1.5 * pdfa.width, wave - pdf.width(wave))
width = (max_q - mean_q)

x = np.linspace(mean_q - 1* width, mean_q + 1 * width, 50001)

plt.plot(x, rk.pq_wavelength(pdf.crossing, 0.65, wave, x), label='crossing');
plt.plot(x, rk.pq_wavelength(pdf.burst, 0.65, wave, x), label='burst');
plt.plot(x, rk.pq_wavelength(pdf.da, 0.65, wave, x), label='rebin');
plt.plot(x, rk.pq_theta(pdfa, 0.65, wave, x), label='angular')
plt.legend();
```

```python
q, k = rk.resolution_kernel(pdfa, pdf, 0.65, wave, npnts=1001)
plt.plot(q, k, label='kernel')

dt = general.div(1.611, 0.167)[0] / 0.65
dl = general.resolution_double_chopper(wave, z0=0.358, R=0.35, freq=24,
                              H=0.004, L=7.5, tau_da=0)
dq = general.q(0.65, wave) * np.sqrt(dl**2 + dt**2) / 2.35

plt.plot(x, stats.norm.pdf(x, loc=mean_q, scale=dq), label='gapprox')
plt.legend()
plt.plot([mean_q - 3 * dq] * 2, [1e-10, 3000])
# plt.yscale('log')
```
