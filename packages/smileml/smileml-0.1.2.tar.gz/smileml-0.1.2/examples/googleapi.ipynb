```python
from dodoml.misc.googleapi import (
    get_google_service,
    walk_google_drive,
    get_analytics_results)
```

# Google drive

```python
# Get connection
scope = ['https://www.googleapis.com/auth/drive']
service = get_google_service('drive', 'v3', scope, '/Users/phi/client_secrets.json')
```

```python
# List all files
service.files().list().execute()
```

```python
# Walk
i = 0
for path, root, dirs, files in walk_google_drive(service, 'LJN'):
    print('%s   %d %d' % ('/'.join(path) + '/' + root['name'], len(dirs), len(files)))
    i += 1
    if i > 5:
        break
```
